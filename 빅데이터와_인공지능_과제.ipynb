```python
# 라이브러리 import
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sb
plt.rc('font', family='NanumBarunGothic')

# url 입력
url = 'https://api.odcloud.kr/api/15004246/v1/uddi:6c4f2ce3-5a37-4c02-969d-9651e9b90188_201712151547?page=1&perPage=10&serviceKey='
key = '4QBrIhMaSOah3vyhwfqOgMkMn87c6%2FcfJBlv%2BfCJUwThou99kcSqkUuZ%2F4sKmpI%2BUVA7%2Fw%2Fd3dUWNn9t6UqEtQ%3D%3D'

# url 불러오기
response = requests.get(url+key)

#데이터 값 출력
contents = response.text
print(contents)

```

```python
#문자열을 json으로 변경
import json
json_ob = json.loads(contents)
print(json_ob)
print(json_ob.keys())
print(json_ob.values())

#key를 기준으로 필요한 내용 추출
count = json_ob['currentCount']
print(count)

data = json_ob['data']
print(data)
```

```python
# key를 기준으로 필요한 내용 추출후 데이터프레임으로 변경
import pandas as pd
data_df = pd.DataFrame(json_ob['data'])
print(data_df)
data_df.to_csv('국토교통부_표준지공시지가_0_9.csv')

```

```python
df = pd.read_csv('국토교통부_표준지공시지가_0_9.csv')
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df
```

```python
print('공시지가 최고값->', df['공시지가'].max())
print('공시지가 최저값->', df['공시지가'].min())
print('공시지가 평균값->', df['공시지가'].mean())
print('공시지가 표준편차->', df['공시지가'].std())
```

```python
plt.hist(df['형상명'])
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (12,6)
sb.regplot(x='면적', y='공시지가', data = df)
plt.xlim(df['면적'].min()-300, df['면적'].max()+300)
plt.grid()
plt.show()
```
