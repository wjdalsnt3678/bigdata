```python
key1 = ['엄마','아빠','나','동생']
val1 = [5,3,1,2]
key2 = ['삼촌', '나', '동생', '이모']
val2 = [3,4,6,3]

import pandas as pd
df1 = pd.DataFrame([key1, val1]).T
df1.columns = ['key','사과']
df1 = df1.set_index('key')
print(df1)

df2 = pd.DataFrame([key2,val2]).T
df2.columns = ['key', '바나나']
df2 = df2.set_index('key')
print(df2)
```

```python
df_ = pd.concat([df1,df2], axis=1)
print(df_)
```

```python
df_ = pd.concat([df1,df2], axis=0)
print(df_.groupby(df_.index).sum())
```

```python
df3 = pd.DataFrame([key2,val2]).T
df3.columns = ['key', '사과']
df3 = df3.set_index('key')
print(df3)
```

```python
df = pd.concat([df1,df3], axis=0)
print(df)
```

```python
df = pd.concat([df1,df3], axis=0)
print(df.groupby(df.index).sum())
```
